### Activity 1: Identify the Components

*1.1. What are the inputs?*
* The marks for three quizzes (quiz_1, quiz_2, quiz_3).
* The user's choice to repeat the process (choice).

*1.2. What is the process?*
* Calculate the average by adding the three quiz marks and dividing by 3.
* Check if the calculated average is 50 or above to determine if the student passes or fails.
* Repeat the entire process if the user chooses "y".

*1.3. What is the output?*
* The student's average mark along with a Pass/Fail message ("Student passes" or "Student fails").
* A final farewell message ("Program Ended") when the program terminates.

---

### Activity 2: Design the Algorithm

*2.1. Pseudocode*
```text
SET choice = "y"
WHILE choice == "y" DO
    INPUT quiz_1
    INPUT quiz_2
    INPUT quiz_3
    
    COMPUTE average = (quiz_1 + quiz_2 + quiz_3) / 3
    
    IF average >= 50 THEN
        PRINT "Student passes"
    ELSE
        PRINT "Student fails"
    ENDIF
    
    INPUT choice ("Continue? Select Y/N: ")
ENDWHILE

PRINT "Program Ended"
 2.2. Identify the algorithm structure:
 2.2.1. Sequence: The step-by-step sequential flow of taking the quiz inputs, performing the average calculation, and asking for the user's choice at the end.
 2.2.2. Selection: The ⁠IF / ELSE⁠ decision structure used to check if the student's average is greater than or equal to 50 to display the correct pass/fail status.
 2.2.3. Iteration: The ⁠WHILE⁠ loop control structure that repeats the entire block of code as long as the variable ⁠choice⁠ remains equal to "y".
